```python
import pandas as pd
```

```python
df = pd.read_csv('greek2.csv')
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

### 1. Separate Names in english  

```python
tmp = pd.DataFrame(df['Names en'].str.rsplit(" ", 1).tolist(), columns=["Given name", "Surname"])
df = pd.concat([df, tmp], axis=1)
```

```python
df
```

### 2. Separates names in greek 

```python
tmp = pd.DataFrame(df['Names gr'].str.rsplit(" ", 1).tolist(), columns=["Greek First Name", "Greek Last Name"])
df = pd.concat([df, tmp], axis=1)
```

```python
df
```

```python
df = df.drop(['Names gr', 'Names en'], axis=1)
```

```python
df
```

### 3. Save the file

```python
# set csv output file_name
file_name = 'greek3.csv'
df.to_csv(file_name, encoding='utf-8')
```

# Format the rest names

```python
df2 = pd.read_csv('greek_rest.csv')
```

### 4. Separate Names in greek 

```python
tmp4 = pd.DataFrame(df2['Names gr'].str.rsplit(" ", 1).tolist(), columns=["Greek First Name", "Greek Last Name"])
df3 = pd.concat([df2, tmp4], axis=1)
```

```python
df3
```

### 5. Separate Names in Enlgish

```python
# strip white space
df3['Names en'] = df3['Names en'].str.strip()
```

```python
tmp5 = pd.DataFrame(df3['Names en'].str.rsplit(" ", 1).tolist(), columns=["Given name", "Surname"])
df4 = pd.concat([df3, tmp5], axis=1)
```

```python
df4 = df4.drop(['Names en', 'Names gr', 'Unnamed: 0'], axis=1)
```

```python
df4
```

### 6. Save the rest names file

```python
# set csv output file_name
file_name = 'greek_rest2.csv'
df4.to_csv(file_name, encoding='utf-8')
```
